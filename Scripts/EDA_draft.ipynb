```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from utils import datetime_plot
```

```python
csv_path = '../Data/CSVs/removed_Info_merged_Text_df.csv'
df = pd.read_csv(csv_path, index_col=0)

df.head(3)
```

```python
df = df[['Text', 'Subject']]
df.head(3)
```

```python
df2 = df.groupby(['Subject']).Text.agg([len])
```

```python
# Set the colors and the style of the figure
sns.set_style("whitegrid")

# Set the width and height of the figure
plt.figure(figsize=(15,35))

# Add title
plt.title('#user writings, by User')

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=df2.len, y=df2.index, palette='brg', orient='h')

# Add label for vertical axis
plt.xlabel('Number of writings')

plt.show()
```
